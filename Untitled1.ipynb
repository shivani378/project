```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load dataset
df = pd.read_csv("/content/movie_ratings.csv")
df.info()
```

```python
# -----------------------------
# 1. DATA PRE-PROCESSING
# -----------------------------

# Check for missing values
print("Missing values:")
print(df.isnull().sum())
```

```python
# Convert date column to datetime format
df['watch_date'] = pd.to_datetime(df['watch_date'])
# Drop duplicates if any
df.drop_duplicates(inplace=True)
# Handle outliers in rating
z_scores = np.abs(stats.zscore(df['rating']))
df = df[z_scores < 3]
```

```python
# -----------------------------
# 2. EXPLORATORY DATA ANALYSIS
# -----------------------------

# Statistical Overview
print("\n--- Dataset Description ---")
print(df.describe())

print("\n--- Categorical Value Counts ---")
print(df['genre'].value_counts())
print(df['platform'].value_counts())
```

```python
# Graphical EDA
sns.set(style="whitegrid")
# Distribution of Ratings
plt.figure(figsize=(8, 5))
sns.histplot(df['rating'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
# Genre Popularity
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='genre', order=df['genre'].value_counts().index, palette='pastel')
plt.title("Genre Distribution")
plt.xticks(rotation=45)
plt.ylabel("Count")
plt.show()
```

```python
# Platform usage
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='platform', order=df['platform'].value_counts().index, palette='Set2')
plt.title("Platform Usage")
plt.ylabel("Count")
plt.show()
```

```python
# Rating by Genre
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='genre', y='rating', palette='muted')
plt.title("Ratings by Genre")
plt.xticks(rotation=45)
plt.show()
```

```python
# Time series: Ratings over time
monthly_avg = df.resample('M', on='watch_date')['rating'].mean()
plt.figure(figsize=(12, 5))
monthly_avg.plot(marker='o', linestyle='-')
plt.title("Average Ratings Over Time")
plt.xlabel("Date")
plt.ylabel("Average Rating")
plt.grid(True)
plt.show()
```

```python
# 1. Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 2. Load dataset
df = pd.read_csv("/content/movie_ratings.csv")   # Change path if needed

# 3. Data Preparation
# Fix timestamp conversion without warnings
df['timestamp'] = pd.to_datetime(df['watch_date']).astype('int64') // 10**9

# Create new features: year and month
df['year'] = pd.to_datetime(df['watch_date']).dt.year
df['month'] = pd.to_datetime(df['watch_date']).dt.month

# Create target column: 1 if rating > 3, else 0
df['HighRating'] = df['rating'].apply(lambda x: 1 if x > 3 else 0)

# 4. Features and Target
# Selecting better features now
X = df[['timestamp', 'year', 'month']]
y = df['HighRating']

# 5. Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 6. Model Training
model = LogisticRegression(max_iter=1000)  # increased iterations for better convergence
model.fit(X_train, y_train)

# 7. Predictions
y_pred = model.predict(X_test)

# 8. Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
df = pd.read_csv("/content/movie_ratings.csv")

def find_movie_details(movie_name, df):
    # Step 1: Normalize movie titles
    df['movie_title_normalized'] = df['movie_title'].str.lower().str.strip()

    # Choices: Only movie names > 4 characters
    choices = [title for title in df['movie_title_normalized'].dropna().unique().tolist() if len(title) > 4]

    # Normalize input
    movie_name = movie_name.lower().strip()

    # Step 2: Find best match using token_sort_ratio
    best_match, score = process.extractOne(movie_name, choices, scorer=fuzz.token_sort_ratio)

    # Threshold minimum score (say 70)
    if score < 70:
        print("❌ Sorry, no good match found. Please check spelling and try again.")
        return

    # Step 3: Get original movie rows
    matched_movies = df[df['movie_title_normalized'] == best_match]

    if matched_movies.empty:
        print("❌ No matching movie found in the database.")
        return

    # Step 4: Pick top rated if multiple
    top_movie = matched_movies.sort_values(by='rating', ascending=False).iloc[0]

    print(f"\n🔍 Best match found: {top_movie.get('movie_title', 'Unknown')} (Match Score: {score})")
    print("\n🎬 Movie Details:")
    print(f"Title        : {top_movie.get('movie_title', 'Unknown')}")
    print(f"Rating       : {top_movie.get('rating', 'Unknown')}")
    print(f"Release Year : {top_movie.get('release_year', 'Unknown')}")
    if 'platform' in top_movie.index:
        print(f"Platform     : {top_movie.get('platform', 'Unknown')}")
    if 'genre' in top_movie.index:
        print(f"Genre        : {top_movie.get('genre', 'Unknown')}")

# Example run
movie_name_input = input("Enter Movie Name: ")
find_movie_details(movie_name_input, df)
```
